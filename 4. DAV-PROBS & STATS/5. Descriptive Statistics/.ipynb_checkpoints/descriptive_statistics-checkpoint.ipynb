```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sehwag = pd.read_csv("sehwag.csv")
dravid = pd.read_csv("dravid.csv")
```

```python
dravid.head()
```

Compute outliers for Sehwag

```python
p_25 = np.percentile(sehwag["Runs"], 25) # 25th percentile or Q1
p_25
```

```python
p_50 = np.percentile(sehwag["Runs"], 50) # 50th percentile or Q2, "median"
p_50
```

```python
p_75 = np.percentile(sehwag["Runs"], 75) # 75th percentile or Q3
p_75
```

```python
iqr = p_75 - p_25
iqr
```

```python
lower = max(p_25 - 1.5 * iqr, 0)
lower
```

```python
upper = p_75 + 1.5 * iqr
upper
```

```python
sns.boxplot(data=sehwag["Runs"], orient="h")
```

Typically, Sehwag scores 0 to 103
If score is greater than 103, we say that it is an "outlier"

```python
len(sehwag)
```

```python
sehwag_typical = sehwag[(sehwag["Runs"] >= lower) & (sehwag["Runs"]<=upper)]
len(sehwag_typical)
```

```python
sehwag_outlier = sehwag[(sehwag["Runs"] < lower) | (sehwag["Runs"]>upper)]
len(sehwag_outlier)
```

```python
len(sehwag_outlier) / len(sehwag)
```

```python
p_25 = dravid["Runs"].quantile(0.25) # Q1 or p_25
p_50 = dravid["Runs"].quantile(0.5)  # Q2 or p_50 or median
p_75 = dravid["Runs"].quantile(0.75) # Q3 or p_75
print(p_25, p_50, p_75)
```

```python
iqr = p_75 - p_25 
lower = max(p_25 - 1.5*iqr, 0)
upper = p_75 + 1.5*iqr
print(lower, upper)
print(iqr)
```

```python
len(dravid)
```

```python
dravid_typical = dravid[(dravid["Runs"] >= lower) & (dravid["Runs"] <= upper)] # Removed outliers
len(dravid_typical)
```

```python
dravid_outlier = dravid[(dravid["Runs"] < lower) | (dravid["Runs"] > upper)]
len(dravid_outlier)
```

```python
len(dravid_outlier) / len(dravid)
```

# CDF

Cumulative distribution function

```python
df_hw = pd.read_csv("weight-height.csv")
df_hw
```

```python
df_height = df_hw["Height"]
df_height.head()
```

```python
min_height = df_height.min()
min_height
```

```python
max_height = df_height.max()
max_height
```

```python
total = len(df_height)
total
```

```python
x_values = np.linspace(min_height, max_height, 1000)
y_values = []
```

```python
for x in x_values:
    people_shorter_than_x = df_height[df_height<=x]
    num_people_shorter_than_x = len(people_shorter_than_x)
    frac_people_shorter_than_x = num_people_shorter_than_x / total
    y_values.append(frac_people_shorter_than_x)
```

```python
plt.plot(x_values, y_values)
```
