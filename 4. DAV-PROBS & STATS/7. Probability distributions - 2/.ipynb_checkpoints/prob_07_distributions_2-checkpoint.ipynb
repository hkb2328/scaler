```python
from scipy.stats import norm # norm stands for Normal or Gaussian
```

# Mean 65, std dev 2.5

```python
# Fraction of people shorter than 69.1
z = (69.1 - 65)/2.5
norm.cdf(z)
```

```python
# Fraction of people shorter than 67.5
z = (67.5 - 65)/(2.5)
norm.cdf(z)
```

```python
norm.cdf(1)
```

```python
norm.ppf(0.8413447460685429)
```

# Empirical rule 68, 95, 99

```python
norm.cdf(1) - norm.cdf(-1)
```

```python
norm.cdf(2) - norm.cdf(-2)
```

```python
norm.cdf(3) - norm.cdf(-3)
```

```python
norm.cdf(0)
```

# Skater example

```python
# ppf -> percentile (point percent function)
norm.ppf(0.05)
```

```python
norm.ppf(norm.cdf(0.7))
```

```python
norm.cdf(norm.ppf(0.8))
```

```python
norm.cdf(-2)
```
