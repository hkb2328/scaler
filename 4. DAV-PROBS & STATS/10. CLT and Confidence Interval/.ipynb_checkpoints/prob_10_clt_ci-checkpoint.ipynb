```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm
```

# Sehwag

```python
sehwag = pd.read_csv("sehwag.csv")
```

```python
sehwag.head()
```

```python
len(sehwag)
```

## Population mean and population std dev

```python
mu_sehwag = sehwag["Runs"].mean().round(2)
print("Sehwag population mean:", mu_sehwag)
```

```python
sigma_sehwag = sehwag["Runs"].std().round(2)
print("Sehwag population std dev:", sigma_sehwag)
```

## Sample mean

```python
num_samples = 20
```

```python
sehwag["Runs"].sample(num_samples)
```

```python
num_samples = 10
sehwag_samples = sehwag["Runs"].sample(num_samples)
print("Sample mean:", sehwag_samples.mean())
```

```python
num_samples = 70
sehwag_samples = sehwag["Runs"].sample(num_samples)
print("Sample mean:", sehwag_samples.mean())
```

## Sample mean trend

1) As "n" increases, we expect lesser variance<br>
2) If "n" increases a lot, the sample mean will be close to the population mean

```python
sample_mean_trend = []
for person in range(20):
    for num_samples in range(5, len(sehwag)):
        sample = sehwag["Runs"].sample(num_samples)
        sample_mean = np.mean(sample)
        sample_mean_trend.append(sample_mean)
    plt.plot(sample_mean_trend)
    sample_mean_trend = []
plt.xlabel("Number of samples")
plt.ylabel("Sample mean")
```

## Sample mean histogram

```python
# Central limit theorem: Sample means follow the Gaussian distribution
num_samples = 30
collect_sample_means = []
for person in range(10000):
    sample_mean = sehwag["Runs"].sample(num_samples).mean()
    collect_sample_means.append(sample_mean)
sns.histplot(collect_sample_means)
```

Standard error = population std dev / sqrt(n)

```python
m = sehwag["Runs"].sample(num_samples).mean()
```

The Z-Scores corresponding to 0.975 and 0.025 can be computed using the code below

```python
norm.ppf(0.975)
```

```python
norm.ppf(0.025)
```

```python
left = m - 1.96 * sigma_sehwag / np.sqrt(num_samples)
left
```

```python
right = m + 1.96 * sigma_sehwag / np.sqrt(num_samples)
right
```

```python
print(f"95% confidence that the population mean is in {left}, {right}")
```

```python
fig, ax = plt.subplots()
sns.histplot(collect_sample_means)
plt.axvline(left)
plt.axvline(right)
```

# SDE-2 Salary

```python
n = 6
```

```python
survey_1 = [35, 36, 33, 37, 34, 35]
np.mean(survey_1)
```

```python
survey_2 = [20, 37, 17, 50, 53, 33]
np.mean(survey_2)
```

```python
bootstrapped_samples = np.random.choice(survey_1, size=n)
print(bootstrapped_samples)
np.mean(bootstrapped_samples)
```

```python
bootstrapped_means_survey_1 = []
for reps in range(10000):
    bootstrapped_samples = np.random.choice(survey_1, size=6)
    bootstrapped_mean = np.mean(bootstrapped_samples)
    bootstrapped_means_survey_1.append(bootstrapped_mean)
```

```python
sns.histplot(bootstrapped_means_survey_1)
```

```python
upper_1 = np.percentile(bootstrapped_means_survey_1, 97.5)
upper_1
```

```python
lower_1 = np.percentile(bootstrapped_means_survey_1, 2.5)
lower_1
```

```python
bootstrapped_means_survey_2 = []
for reps in range(10000):
    bootstrapped_samples = np.random.choice(survey_2, size=6)
    bootstrapped_mean = np.mean(bootstrapped_samples)
    bootstrapped_means_survey_2.append(bootstrapped_mean)
```

```python
sns.histplot(bootstrapped_means_survey_2)
```

```python
upper_2 = np.percentile(bootstrapped_means_survey_2, 97.5)
upper_2
```

```python
lower_2 = np.percentile(bootstrapped_means_survey_2, 2.5)
lower_2
```
