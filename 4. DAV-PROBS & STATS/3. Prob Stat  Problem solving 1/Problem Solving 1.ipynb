```python
numerator = 0
denominator = 0

num_families = 10000
for family in range(num_families):
    child1 = np.random.choice([0, 1])
    child2 = np.random.choice([0, 1])
    
    ## Let y represent the number of girls
    y = child1 + child2
    if y >= 1:
        denominator += 1
        if y == 2:
            numerator += 1

conditional_prob = numerator/denominator
print(conditional_prob)
```
