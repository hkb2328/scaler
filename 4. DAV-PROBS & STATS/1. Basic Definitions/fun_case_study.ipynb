## Sachin Tendulkar's ODIs for India
(Source: [cricinfo](https://stats.espncricinfo.com/ci/engine/player/35320.html?class=2;template=results;type=batting;view=innings))

#### Imports

```python
import numpy as np
import pandas as pd
```

#### Read the data

```python
df_sachin = pd.read_csv("Sachin_ODI.csv")
df_sachin.head()
```

#### What is the probability of India winning when Sachin is playing?

```python
num_won = len(df_sachin[df_sachin["Won"] == True])
num_matches = len(df_sachin)

p_win = num_won / num_matches

print(f"Prob that india wins when sachin plays: {p_win}")
```

#### What is the probability of Sachin scoring a century?

```python
num_century = len(df_sachin[df_sachin["century"] == True])
num_matches = len(df_sachin)

p_century = num_century / num_matches

print(f"Prob that sachin scores a century: {p_century}")
```

#### What is the probability of Sachin scoring a century and India winning?

```python
num_century_and_win = len(df_sachin[ (df_sachin["Won"] == True) & (df_sachin["century"] == True)])
num_matches = len(df_sachin)

p_won_and_century = num_century_and_win / num_matches

print(f"Prob that sachin scores a century and india wins: {p_won_and_century}")
```

```python
num_century_and_win = len(df_sachin[ (df_sachin["Won"] == False) & (df_sachin["century"] == True)])
num_matches = len(df_sachin)

p_won_and_century = num_century_and_win / num_matches

print(f"Prob that sachin scores a century and india wins: {p_won_and_century}")
```
