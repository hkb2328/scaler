Suppose the store has **beginning on-hand (BOH)** inventory of 1300. If the demand is more than the BOH, there is a need for stock **replenishment**. 

Suppose there are 2000 stores, each with BOH of 1300. The **distribution center (DC)** needs to calculate the number of stores which might need replenishment before next Sunday. Which distribution best characterizes this number?

```python
# P(X > 1300)
z = (1300 - mean) / std
p = 1 - norm.cdf(z)

print(p)
```

```python
n = 2000

print("Expected number of stores which need replenishment: ", binom.expect(args=(n, p)))
```

## Pooled Blood Test

```python
n = 4
p = 0.1
p_x_1 = binom.pmf(k=0, n=n, p=p)
p_x_5 = 1 - p_x_1

expectation = 1 * p_x_1 + 5 * p_x_5

print("Expected number of tests: ", expectation)
```

## Converting Random Variables

```python
## Given a dice - S = {1, 2, 3, 4, 5 ,6}
s = [1, 2, 3, 4, 5 ,6]
def dice_roll():
    return np.random.choice(s)
```

```python
## Coin - S = {H, T}
# P(H) = 0.5
# P(T) = 0.5
def coin_toss():
    d = dice_roll()
    if d > 3:
        return "H"
    else:
        return "T"
```

```python
n = 10000

tosses = []

for i in range(n):
    tosses.append(coin_toss())

tosses = np.array(tosses)
num_heads = len(tosses[tosses=="H"])

prob_heads = num_heads / n

print(prob_heads)
```

```python
### Given a coin, simulate a dice
def three_ct():
    return np.random.choice([0, 1, 2, 3, 4, 5, 6, 7])

def dice_from_ct():
    ct = three_ct()
    if ct > 0 and ct < 7:
        return ct
    else:
        return dice_from_ct()
```

```python
n = 100000

rolls = []

for i in range(n):
    rolls.append(dice_from_ct())

rolls = np.array(rolls)
num_1 = len(rolls[rolls==1])
num_2 = len(rolls[rolls==2])
num_3 = len(rolls[rolls==3])
num_4 = len(rolls[rolls==4])
num_5 = len(rolls[rolls==5])
num_6 = len(rolls[rolls==6])

prob_1 = num_1 / n
prob_2 = num_2 / n
prob_3 = num_3 / n
prob_4 = num_4 / n
prob_5 = num_5 / n
prob_6 = num_6 / n

print(prob_1)
print(prob_2)
print(prob_3)
print(prob_4)
print(prob_5)
print(prob_6)

sns.countplot(rolls)
```
